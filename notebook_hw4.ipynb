```python
import pandas as pd
import hw4_library.read_data.csvreader as rd
import hw4_library.data_preprocessor.preprocessor as pre
import hw4_library.features.transformator as tr
import hw4_library.model.train_Model as trmd
import joblib
```

```python

reader = rd.CSVReader()
train_data, test_data = reader.read('sample_diabetes_mellitus_data.csv')  
```

```python
# Remove NaN
rmNaN = pre.RemoveNaNsRowPreprocessor()
train_data = rmNaN.preprocess(train_data,['age', 'gender', 'ethnicity'])


rmNaN = pre.RemoveNaNsRowPreprocessor()
test_data = rmNaN.preprocess(test_data,['age', 'gender', 'ethnicity'])
test_data.head()
test_data.tail()
```

```python
# Mean fill
fillNaN = pre.FillNaNsColumnPreprocessor()
train_data = fillNaN.preprocess(train_data,['height', 'weight'])

fillNaN = pre.FillNaNsColumnPreprocessor()
test_data = fillNaN.preprocess(test_data,['height', 'weight'])
test_data.head()
test_data.tail()
```

```python
# Transform columns df

# Create instances of the transformation classes
normalizator = tr.NormalizationTransformator()
standardizator = tr.StandarizationTransformator()

# Specify the columns you want to transform
columns_to_transform = ['height', 'weight']  # Adjust as needed

# Apply normalization to both training and test data
train_data = normalizator.transform(train_data, columns_to_transform)
test_data = normalizator.transform(test_data, columns_to_transform)

# Apply standardization to both training and test data
train_data = standardizator.transform(train_data, columns_to_transform)
test_data = standardizator.transform(test_data, columns_to_transform)

```

```python
# Train model, predict, report accuracy score
feature_columns = ['age',
               'height',
               'weight',
               'aids',
               'cirrhosis',
               'hepatic_failure',
               'immunosuppression',
               'leukemia',
               'lymphoma',
               'solid_tumor_with_metastasis']
target_column = 'diabetes_mellitus'


diabetes_model = trmd.Train_Model(feature_columns, target_column)
diabetes_model.train(train_data)
test_data['predictions'] = diabetes_model.predict(test_data)

diabetes_model.accuracy(test_data, 'predictions')
```
